# Merge Data SST Pasuruan (2016-2025)

Notebook ini menggabungkan seluruh file CSV tahunan SST (Sea Surface Temperature) untuk wilayah Pasuruan (2016-2025) menjadi **satu file CSV** utuh.

**Tahapan:**
1. Import library yang dibutuhkan
2. Cari semua file `.csv` di folder data
3. Baca & gabungkan (merge) seluruh file menjadi satu DataFrame
4. Bersihkan & rapikan data
5. Export ke CSV


```python
# 1. Import library
import os
import glob
import pandas as pd
```

```python
# 2. Cari semua file .csv di folder data
data_folder = "data"  # folder tempat semua file csv per tahun disimpan
file_list = sorted(glob.glob(os.path.join(data_folder, "sst_pasuruan_*.csv")))

print(f"Jumlah file ditemukan: {len(file_list)}")
file_list
```

## 3. Gabungkan seluruh file CSV

Setiap file memiliki baris kedua berisi satuan (unit), bukan data, sehingga baris tersebut perlu dilewati (`skiprows=[1]`) saat dibaca.


```python
# 3. Baca dan gabungkan semua file csv menjadi satu DataFrame
df_list = []
for f in file_list:
    temp = pd.read_csv(f, skiprows=[1])  # baris ke-2 berisi satuan, dilewati
    df_list.append(temp)

df = pd.concat(df_list, ignore_index=True)
print(f"Jumlah baris setelah digabung: {len(df)}")
df.head()
```

## 4. Bersihkan & rapikan data

- Ubah kolom `time` menjadi format datetime
- Urutkan berdasarkan waktu
- Hilangkan duplikat (jika ada)


```python
# 4. Bersihkan & rapikan data
df["time"] = pd.to_datetime(df["time"])
df = df.rename(columns={"time": "datetime"})

df = df.sort_values("datetime").drop_duplicates().reset_index(drop=True)

print("Rentang waktu data:", df["datetime"].min(), "sampai", df["datetime"].max())
print("Jumlah baris akhir:", len(df))
df.head()
```

## 5. Export ke CSV

Simpan seluruh data yang sudah digabung ke dalam satu file CSV utuh.


```python
# 5. Simpan ke file CSV
output_path = "sst_pasuruan_2016-2025_merged.csv"
df.to_csv(output_path, index=False)

print(f"Berhasil disimpan: {output_path}")
print(f"Total baris: {len(df):,}")
print(f"Total kolom: {len(df.columns)}")
```
